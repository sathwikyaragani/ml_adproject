```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the dataset manually or load it from a CSV file
data = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Product_Name': ['PainRelief', 'ColdCure', 'HeadacheFix', 'PainRelief', 'ColdCure', 'HeadacheFix', 'PainRelief', 'ColdCure', 'HeadacheFix', 'PainRelief'],
    'Region': ['North America', 'Europe', 'Asia', 'North America', 'Europe', 'Asia', 'North America', 'Europe', 'Asia', 'North America'],
    'Marketing_Spend': [50000, 30000, 45000, 52000, 28000, 48000, 51000, 31000, 47000, 53000],
    'Sales': [100000, 60000, 80000, 110000, 55000, 90000, 105000, 65000, 85000, 115000],
    'Effectiveness': [85, 75, 80, 88, 70, 82, 86, 78, 80, 90],
    'Side_Effects': [5, 2, 3, 4, 1, 3, 5, 2, 4, 6],
    'Age_Group': ['20-30', '30-40', '40-50', '50-60', '30-40', '40-50', '20-30', '30-40', '50-60', '20-30'],
    'Trial_Period': ['6 months', '3 months', '6 months', '12 months', '3 months', '6 months', '6 months', '3 months', '12 months', '6 months']
})

# Check the first few rows of the dataset
print(data.head())

# Data Cleaning (checking for missing values and duplicates)
print("\nMissing values in the dataset:\n", data.isnull().sum())
data = data.dropna()  # Drop rows with missing values
print("\nDuplicate rows in the dataset: ", data.duplicated().sum())
data = data.drop_duplicates()

# ---- Visualizations ----

# 1. Bar plot comparing the average Effectiveness for each drug across different regions
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Region', y='Effectiveness', hue='Product_Name')
plt.title('Average Effectiveness of Each Drug Across Regions')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Violin plot to show the distribution of Effectiveness for each product
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='Product_Name', y='Effectiveness', inner='quart', palette='muted')
plt.title('Distribution of Effectiveness by Drug')
plt.xlabel('Drug')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# Violin plot for Side Effects
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='Product_Name', y='Side_Effects', inner='quart', palette='muted')
plt.title('Distribution of Side Effects by Drug')
plt.xlabel('Drug')
plt.ylabel('Side Effects')
plt.tight_layout()
plt.show()

# 3. Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(data[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.suptitle('Pairplot of Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.tight_layout()
plt.show()

# 4. Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Trial_Period', y='Effectiveness')
plt.title('Effectiveness of Drugs Across Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# 5. Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(data=data, x='Marketing_Spend', y='Effectiveness', scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Marketing Spend vs Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()

# ---- Data Analysis ----

# a. Which product has the best overall effectiveness?
avg_effectiveness = data.groupby('Product_Name')['Effectiveness'].mean().sort_values(ascending=False)
print("\nAverage Effectiveness for Each Drug:\n", avg_effectiveness)

# b. The correlation between effectiveness and side effects
correlation_matrix = data[['Effectiveness', 'Side_Effects']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Effectiveness and Side Effects')
plt.show()
```
